```python
import pygame
import sys

pygame.init()
pygame.joystick.init()

if pygame.joystick.get_count() == 0:
    print("No joystick detected. Please connect a controller and re-run.")
    sys.exit(1)

joystick = pygame.joystick.Joystick(0)
joystick.init()

print("Joystick initialized:", joystick.get_name())
print("Number of axes:", joystick.get_numaxes())

running = True
clock = pygame.time.Clock()

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # Optional: Press 'ESC' key to quit
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                running = False

    # Clear the screen by printing a separator line (to keep console output readable)
    print("-" * 30)

    # Print out the values of all axes
    num_axes = joystick.get_numaxes()
    for i in range(num_axes):
        axis_value = joystick.get_axis(i)
        print(f"Axis {i}: {axis_value:.3f}")

    # After identifying the L2 axis and its range, you can try thresholds here
    # For example, if L2 is on axis 2, and resting around 0.0 and pressed around 1.0:
    # l2_value = joystick.get_axis(2)
    # l2_pressed = l2_value > 0.5
    # print(f"L2 Pressed: {l2_pressed}")

    clock.tick(10)  # Limit updates to 10 times per second for readability

pygame.quit()
```
